```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.head(2)
```

```python
titanic.drop('PassengerId',axis=1, inplace=True)
titanic.head(2)
```

```python
titanic.groupby(titanic['Age'].isnull()).mean()
```

```python
titanic['Age'].fillna(titanic['Age'].mean(), inplace=True)
titanic.isnull().sum()
```

```python
titanic.head(10)
```

```python
titanic['family'] = titanic['SibSp'] + titanic['Parch']
titanic.head()
```

```python
titanic.drop(['Pclass','Name','Age','SibSp','Parch','Fare'], axis=1, inplace=True)
titanic.head()
```

```python
titanic.info()
```

```python
titanic.groupby(titanic['Cabin'].isnull()).mean()
```

```python
# criando um estado usando where onde o primeiro argumeto e a coluna qu queremos
# o segundo argumento e o valor que queremos atribuir para a nova coluna se for verdade
# o terceiro argumento e o valor que queremos atribuir para a nova coluna se for falso
titanic['Cabin_ind'] = np.where(titanic['Cabin'].isnull(), 0, 1)
titanic.head(10)
```

```python
# plot categorical features
for index, col in enumerate(['Cabin_ind', 'Sex', 'Embarked']):
    plt.figure(index)
    sns.catplot(x=col, y='Survived', data=titanic, kind='point', aspect=3)
```

```python
titanic.pivot_table('Survived', index='Sex', columns='Embarked', aggfunc='count')
```

```python
titanic.pivot_table('Survived', index='Cabin_ind', columns='Embarked', aggfunc='count')
```

```python
# Convert Sex to numeric ass male = 0 and female = 1
# usando o metodo map para remapear os valores das variaveis
gender_num = {'male':0, 'female':1 }
titanic['Sex'] = titanic['Sex'].map(gender_num)
titanic.head()
```

```python
titanic.drop(['Cabin', 'Embarked'], axis=1, inplace=True)
titanic.head()
```
